```python
import IPython.display as idp
import librosa
```

## Dataset audio files sampled at 22k and 8k 

Note that whilst the drilling is very clear at 22k, it is almost unrecognizable at 8k.

### Sample 1

```python
audio, rt = librosa.load('data/audio/fold5/17578-5-0-1.wav', sr=22050, mono=True,  res_type='kaiser_fast')
idp.Audio(audio, rate=rt)
```

```python
audio, rt = librosa.load('data/audio/fold5/17578-5-0-1.wav', sr=8000, mono=True,  res_type='kaiser_fast')
idp.Audio(audio, rate=rt)
```
